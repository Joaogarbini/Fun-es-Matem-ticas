Exemplo 4- Resolva e represente graficamente cada sistema de equações lineares.

 a) x + y = 3        b) x + y = 3         c) x + y = 3
    x - y = -1          2x + 2y = 6          x + y = 1


```python
# a)
solve([x + y - 3, x -y +1],[x,y]) 
```

```python
# b) Esse sitema possui infinitas soluções onde o y pode ser qualquer número real
solve([x + y - 3, 2*x + 2*y -6],[x,y]) 
```

```python
#c) Esse sistema não possui nenhuma solução pois a soma de dois números não pode ser 3 e 1 simultaneamente
solve([x +y -3, x + y -1],[x,y])
```

Abaixo temos os gráficos dos 3 sistemas, essas são imagens originais retiradas do livro do Ron Larson 



```python
# Importando o pacote matplotlib e o pylab para reproduzirmos os gráficos da imagem
import matplotlib.pyplot as plt
%matplotlib inline
from pylab import plot, arange
```

```python
#Gráfico do sistema a) o ponto de intersecção das retas é exatamente a solução do sistema x=1, y=2 
x = arange(-1,4,1)
y=3-x
y1=1+x
plot(x,y)
plot(x,y1)
```

```python
# Gráfico do sistema b) as duas reatas são coincidentes uma sobre a outra e diferente do gráfico anterior onde tinhamos apenas
#um ponto de intersecção aqui teremos infinitos pontos de intersecção
x = arange(-1,4,1)
y=3-x
y1=3-x
plot(x,y)
plot(x,y1)
```

```python
#Gráfico do sistema c) sendo as retas paralelas nunca teremos ponto de intersecção, indicando que não há solução para o sistema
x = arange(-1,4,1)
y=3-x
y1=1-x
plot(x,y)
plot(x,y1)
```
